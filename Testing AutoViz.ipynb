# Testing AutoViz Python Package
https://github.com/AutoViML/AutoViz

I saw this package in an article today, and I was interested in checking out whether or not it is a viable option for my data visualizations.

## Imports

```python
from autoviz.AutoViz_Class import AutoViz_Class
import pandas as pd

AV = AutoViz_Class()
```

## Numeric, String-Categorical, & String-Boolean Columns
### Demonstration
Dataset - https://www.kaggle.com/benroshan/factors-affecting-campus-placement

#### Regression

```python
filename = "Placement_Data_Full_Class.csv"
dft = AV.AutoViz(
    filename,
    sep=",",
    depVar="salary",
    dfte=None,
    header=0,
    verbose=1,
    lowess=False,
    chart_format="svg",
    max_rows_analyzed=150000,
    max_cols_analyzed=30,
)
```

#### Classification

```python
dft = AV.AutoViz(
    filename,
    sep=",",
    depVar="status",
    dfte=None,
    header=0,
    verbose=1,
    lowess=False,
    chart_format="svg",
    max_rows_analyzed=150000,
    max_cols_analyzed=30,
)
```

### Comments

I was impressed by how fast this worked and how automated the process is. Clearly it is just putting out a huge number of charts, but the charts do seem to be catered to the problem which is really neat. I think this is great for initial exploratory analysis of a new dataset. Too easy not to at least look at it as an option.
